### Walk Through STADATA python library to call data from BSP Web API

#### Import library

```python
import os
import stadata
import pandas as pd
```

#### Initialization

```python
TOKEN = os.environ.get("TOKEN")
client = stadata.Client(token=TOKEN)
```

#### List domain

```python
list_domain = client.list_domain()
list_domain
```

#### Static table

```python
list_all_static_table = client.list_statictable(all=True)
list_certain_domain_static_table = client.list_statictable(all=False, domain=['1871'])
```

```python
list_all_static_table
```

```python
list_certain_domain_static_table
```

```python
#view ...
view_static_table = client.view_statictable(domain=1800, table_id=2172, lang="ind")
view_static_table
```

#### Dynamic table

```python
list_all_dynamic_table = client.list_dynamictable(all=True)
list_certain_domain_dynamic_table = client.list_dynamictable(all=False, domain=['1800'])
```

```python
list_all_dynamic_table
```

```python
list_certain_domain_dynamic_table
```

#### List Publication

```python
list_publication = client.list_publication(all=True)
list_publication_domain = client.list_publication(all=False, domain=['1800'])
list_publication_spec_time = client.list_publication(all=False, domain=['1800'], month='4', year='2023')
```

```python
list_publication
```

```python
list_publication_domain
```

```python
list_publication_spec_time
```

#### List Press Release

```python
# Get all static tables from all domains
list_press = client.list_pressrelease(all=True)

# Get static tables from specific domains
list_press_domain = client.list_pressrelease(all=False, domain=['1800'])

# Get static tables from specific domains, year, and month
list_press_spec_time = client.list_pressrelease(all=False, domain=['1800'], month="4", year="2023")
```
